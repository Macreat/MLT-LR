## Ej 1 



Sea el modelo de regresión:
$$
t_n = \phi(x_n)^\top w + \eta_n, \quad \eta_n \sim \mathcal{N}(0,\sigma_\eta^2), \quad n=1,\dots,N,
$$
con datos i.i.d. y $\phi:\mathbb{R}^P \to \mathbb{R}^Q$, $Q \ge P$. Presentamos la siguiente comparativa

### MÍNIMOS CUADRADOS (OLS)

### MÍNIMOS CUADRADOS REGULARIZADOS 

### MÁXIMA VEROSIMILITUD (MLE)

### MÁXIMO A POSTERIORI (MAP)

### MODELO BAYESIANO LINEAL GAUSSIANO 

### REGRESIÓN RÍGIDA KERNEL (KRR)

### PROCESOS GAUSIANOS (GP)

### Finalmente discutimos similitudes y diferencias:  

####  Similitudes estructurales



Todos los modelos comparten la misma hipótesis lineal y una función de costo basada en el error cuadrático medio:
$$
J(w) = \|t - \Phi w\|_2^2,
$$
derivada de la asunción de ruido gaussiano.  
La diferencia central radica en el **tratamiento de $w$**:  
- **LS / ML** estiman un único punto $\hat{w}$ determinístico.  
- **MAP / Ridge** introducen regularización (prior gaussiano).  
- **Bayesiano / GP** extienden a una distribución posterior sobre $w$ o $f(x)$.


####  Diferencias conceptuales



| Modelo | Naturaleza | Diferencia clave |
|:--|:--|:--|
| **LS / ML** | Determinista | Mínimos Cuadrados ≡ Máx. Verosimilitud bajo ruido $\mathcal{N}(0,\sigma^2I)$ |
| **Ridge / MAP** | Regularizado | Añaden prior gaussiano $p(w)\!\sim\!\mathcal{N}(0,\lambda^{-1}I)$ → penalización L2 |
| **Bayesiano Lineal** | Probabilístico paramétrico | Propaga incertidumbre sobre $w$ → posterior $p(w|t,X)=\mathcal{N}(m_N,S_N)$ |
| **Kernel Ridge** | Determinista kernelizado | Reexpresa Ridge en espacio de funciones vía kernel $k(x_i,x_j)$ |
| **Procesos Gaussianos (GP)** | Bayesiano no paramétrico | Define un prior sobre funciones $f(x)\!\sim\!\mathcal{GP}(0,k)$ y modela la covarianza predictiva |


####  Conclusiones técnicas



1. **OLS, ML y MAP** comparten la misma estructura cuadrática; difieren solo por la inclusión de priors o términos de regularización.  
2. **Bayesiano y GP** generalizan MAP al estimar distribuciones completas, permitiendo predicciones con incertidumbre.  
3. **Kernel Ridge y GP** son equivalentes en la media predictiva; el GP añade varianza, extendiendo Ridge a un marco bayesiano no paramétrico.

| Regressor | Modelo matemático | Función de costo $J(w)$ | Estrategia de optimización | Relación con punto 1 | Escalabilidad |
|:--|:--|:--|:--|:--|:--|
| **LinearRegression (OLS)** | $\hat{y}=Xw+b$ | $J=\frac{1}{2n}\|y-Xw-b\|_2^2$ | Analítica (solución cerrada) | Equivalente a Mínimos Cuadrados / MLE | Excelente |
| **Lasso (L1 Regularization)** | $\hat{y}=Xw+b$ | $J=\frac{1}{2n}\|y-Xw-b\|_2^2+\alpha\|w\|_1$ | Descenso por coordenadas | MAP Laplace / Sparse solution | Media |
| **Ridge (L2 Regularization)** | $\hat{y}=Xw+b$ | $J=\frac{1}{2n}\|y-Xw-b\|_2^2+\alpha\|w\|_2^2$ | Analítica — resuelve $(X^\top X+\lambda I)w=X^\top y$ | MAP Gaussiano / Regularización L2 | Alta |
| **ElasticNet (L1 + L2)** | $\hat{y}=Xw+b$ | $J=\frac{1}{2n}\|y-Xw-b\|_2^2+\alpha_1\|w\|_1+\alpha_2\|w\|_2^2$ | Descenso por coordenadas / proximal | Interpolación Lasso–Ridge | Media |
| **KernelRidge** | $\hat{y}(x)=\sum_i\alpha_i k(x_i,x)$ | $J=\|y-K\alpha\|_2^2+\lambda\|\alpha\|_2^2$ | Dual (kernel) — analítica | Ridge kernelizado / MAP con kernel $k$ | Baja ($O(N^3)$) |
| **SGDRegressor** | $\hat{y}=Xw+b$ | $J=\frac{1}{2n}\|y-Xw-b\|_2^2$ | Descenso de gradiente estocástico (online / mini-batch) | Aproximación iterativa de OLS | Excelente (escala masiva) |
| **BayesianRidge** | $\hat{y}=Xw+b$ | Posterior $p(w\|X,y)=\mathcal{N}(m_N,S_N)$, $S_N^{-1}=\alpha I+\beta X^\top X$ | Inferencia bayesiana (evidence max.) | Extensión probabilística del MAP Ridge | Alta |
| **GaussianProcessRegressor (GPR)** | $f(x)\sim\mathcal{GP}(0,k(x,x'))$ | $-\log p(y\|X)=\frac{1}{2}y^\top(K+\sigma^2I)^{-1}y+\frac{1}{2}\log\|K+\sigma^2I\|$ | Máx. de evidencia / gradiente de hiperparámetros | Extensión probabilística de Kernel Ridge | Limitada ($O(N^3)$) |
| **Support Vector Regressor (SVR)** | $f(x)=\sum_i(\alpha_i-\alpha_i^*)k(x_i,x)+b$ | $\min_{w,b,\xi,\xi^*}\frac{1}{2}\|w\|^2+C\sum_i(\xi_i+\xi_i^*)$ s.a. $\|y_i-w^\top\phi(x_i)-b\|\leq\epsilon+\xi_i$ | Programación Cuadrática (QP) | Regularización de margen / ε-insensitive loss | Media |
| **RandomForestRegressor** | $\hat{f}(x)=\frac{1}{M}\sum_{m=1}^M h_m(x)$ | $MSE_{nodo}=\frac{1}{N_{nodo}}\sum_i(y_i-\bar{y}_{nodo})^2$ | Bagging + división voraz (greedy split) | Ensamble no paramétrico (bosque de árboles) | Alta (paralela) |
| **GradientBoosting / XGBoost** | $\hat{f}(x)=\sum_{k=1}^K f_k(x)$ | $J=\sum_i L(y_i,\hat{y}_i)+\sum_k\Omega(f_k)$, con $L=\frac{1}{2}(y_i-\hat{y}_i)^2$ y $\Omega=\gamma T+\frac{1}{2}\lambda\sum_jw_j^2$ | Boosting secuencial / gradiente funcional | Ensamble aditivo / descenso en espacio de funciones | Muy alta (GPU-compatible) |

Vemos, como en conjunto, los regresores analizados representan un continuo entre modelos lineales simples y métodos no lineales o probabilísticos más complejos; la elección del modelo depende del compromiso entre interpretabilidad, capacidad de generalización y costo computacional. 

Además, es claro, que los modelos lineales ofrecen rapidez y claridad matemática, mientras que los kernelizados y ensambles amplían la flexibilidad funcional a costa de mayor complejidad; la diversidad de optimizadores y regularizaciones permite adaptar el aprendizaje a distintas escalas y estructuras de datos.

#### Descripción general




**RAPIDS AI** es un ecosistema de librerías aceleradas por **GPU (CUDA)**.  
Permite ejecutar operaciones de *data science* y *machine learning* en GPU con una sintaxis casi idéntica a **pandas** y **scikit-learn**.

**Arquitectura general:**

- **cuDF** → Manipulación de datos en GPU (reemplazo de `pandas`)  
- **cuML** → Algoritmos de *Machine Learning* (reemplazo de `scikit-learn`)  
- **cuGraph** → Algoritmos de grafos sobre GPU  

**Fuente:** [https://rapids.ai](https://rapids.ai)

### preprocesamiento

una vez vista las distribuciones de las variables correlacionadas, podemos realizar una ingeniería de características con el fin de ayudar a la interpretabilidad de los modelos

e.g, es apta la conversión de altura a pulgadas


```python
def parse_height(h):
    if isinstance(h, str) and '-' in h:
        ft, inch = h.split('-')
        return int(ft) * 12 + int(inch)
    return np.nan
```

aplicamos conversión de altura al data set

```python

tracking_df['player_height'] = tracking_df['player_height'].apply(parse_height)
```

de igual manera, podemos calcular edad del jugador (en años) usando una fecha de referencia


```python
tracking_df['player_birth_date'] = pd.to_datetime(tracking_df['player_birth_date'], errors='coerce')
reference_date = pd.to_datetime('2025-10-23')
tracking_df['age'] = (reference_date - tracking_df['player_birth_date']).dt.days / 365.25
```

ahora, procedemos a definir las listas de variables numéricas y categóricas


```python


numeric_features = ['x', 'y', 's', 'a', 'o', 'dir', 'player_weight', 'player_height', 'age', 'absolute_yardline_number']
categorical_features = ['player_position', 'player_side', 'player_role']
```

una vez realizada la distinción de nuestras variables, podemos definir el preprocesador para columnas numéricas y categóricas

comenzando con el preporcesador para las numéricas

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

de igual forma para las categóricas

```python
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

finalmente, obtenemos nuestro preprocesador con ayuda de columnTransformer

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

aqui, procedemos a preparar los datos para nuestros diferentes modelos de aprendizaje

```python
tracking_df.info()
```

comenzando por la definicion de las variables de interés

```python
target_cols = ['x', 'y']
```

de igual manera, tenemos para las variable de trabajo

```python
 
merge_cols = ['game_id', 'play_id', 'nfl_id', 'frame_id']

```

finalmente, realizamos una copia de nuestro dataset

```python

data_full = tracking_df.merge(
    labels_df[merge_cols + target_cols],
    on=merge_cols,
    how='inner',
    suffixes=('', '_target')  # deja las features sin sufijo y marca los targets
)
```

creamos la base para nuestro sample submission solicitado

```python
target_cols = [f'{col}_target' for col in target_cols]  # ['x_target', 'y_target']
X = data_full[numeric_features + categorical_features]
y = data_full[target_cols]
```


podemos separar las características (X) de las etiquetas (y)

```python
X = data_full[numeric_features + categorical_features]
y = data_full[target_cols]
```

ahora, realizamos la particion del conjunto de datos (hold out)

#### kernel ridge

ahora, con el kernel por rejilla

```python
from sklearn.kernel_approximation import Nystroem
from sklearn.linear_model import Ridge
import optuna
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Kernel Ridge Approximation (Nystroem + Ridge) para limitar memoria en 16 GB
def objective_krr_approx(trial):
    alpha = trial.suggest_float('alpha', 1e-3, 1e2, log=True)
    gamma = trial.suggest_float('gamma', 1e-3, 1e1, log=True)
    n_components = trial.suggest_int('n_components', 500, 3000)

    krr_pipeline = Pipeline(steps=[
        ('preprocess', preprocessor),
        ('feature_map', Nystroem(kernel='rbf', gamma=gamma, n_components=n_components, random_state=42)),
        ('reg', Ridge(alpha=alpha))
    ])

    # Entrenar sobre un subconjunto pequeño para evitar saturar memoria
    subset_size = min(10000, len(X_train))
    idx = np.random.choice(len(X_train), subset_size, replace=False)
    X_train_sub, y_train_sub = X_train.iloc[idx], y_train.iloc[idx]

    krr_pipeline.fit(X_train_sub, y_train_sub)
    preds = krr_pipeline.predict(X_val)
    mse = mean_squared_error(y_val, preds)
    return mse
```

creamos el estudio con optuna

```python

study_krr = optuna.create_study(direction='minimize')
study_krr.optimize(objective_krr_approx, n_trials=5, show_progress_bar=False)

best_alpha_krr = study_krr.best_params['alpha']
best_gamma_krr = study_krr.best_params['gamma']
best_ncomp_krr = study_krr.best_params['n_components']
```

de igual forma, procedemos creando el pipeline

```python

krr_best = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('feature_map', Nystroem(kernel='rbf', gamma=best_gamma_krr, n_components=best_ncomp_krr, random_state=42)),
    ('reg', Ridge(alpha=best_alpha_krr))
])
```

realizamos un submuestreo 

```python

subset_size = min(15000, len(X_train))
idx = np.random.choice(len(X_train), subset_size, replace=False)
X_train_sub, y_train_sub = X_train.iloc[idx], y_train.iloc[idx]
```

y finalmente, tenemos el entrenamiento con los mejores hpo

#### SVR

de igual forma para el regresor con maquinas de soporte vectorial

creamos la función objetivo

```python
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import optuna
import numpy as np

# --- Función objetivo (usa subset para evitar bloqueo) ---
def objective_svr(trial):
    # Rango de hiperparámetros con sintaxis moderna (evita FutureWarning)
    C = trial.suggest_float('C', 1e-1, 1e3, log=True)
    epsilon = trial.suggest_float('epsilon', 1e-3, 1.0, log=True)
    gamma = trial.suggest_float('gamma', 1e-3, 1e1, log=True)

    # Submuestreo de entrenamiento para acelerar (ajusta según tu RAM)
    subset_size = min(5000, len(X_train))
    idx = np.random.choice(len(X_train), subset_size, replace=False)
    X_train_sub = X_train.iloc[idx]
    y_train_sub = y_train.iloc[idx, 0]  # solo coord X para Optuna

    svr_pipeline = Pipeline(steps=[
        ('preprocess', preprocessor),
        ('reg', SVR(C=C, epsilon=epsilon, kernel='rbf', gamma=gamma))
    ])
    svr_pipeline.fit(X_train_sub, y_train_sub)

    preds = svr_pipeline.predict(X_val)
    mse = mean_squared_error(y_val.iloc[:, 0], preds)
    return mse
```

y configuramos el estudio de optuna con un límite de intentos::

```python

study_svr = optuna.create_study(direction='minimize')
study_svr.optimize(objective_svr, n_trials=3, show_progress_bar=False)  # pocos trials para que no se bloquee
```

para finalmente obtener los mejores parámetros

procedemos entrenando dos SVR con estos hiperparámetros

Primeramente con un submuestreo para evitar saturar memoria

```python


from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
import numpy as np


subset_train = min(1000, len(X_train))
subset_val   = min(2000, len(X_val))

train_idx = np.random.choice(len(X_train), subset_train, replace=False)
val_idx   = np.random.choice(len(X_val), subset_val, replace=False)

X_train_gp = X_train.iloc[train_idx]
y_train_gp = y_train.iloc[train_idx]
X_val_gp   = X_val.iloc[val_idx]
y_val_gp   = y_val.iloc[val_idx]
```


procedemos con la definición del kernel RBF y dos modelos GP (X e Y)


```python


kernel_gp = ConstantKernel(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))

gp_x = Pipeline([
    ('preprocess', preprocessor),
    ('reg', GaussianProcessRegressor(kernel=kernel_gp, alpha=1e-2, n_restarts_optimizer=2, random_state=42))
])

gp_y = Pipeline([
    ('preprocess', preprocessor),
    ('reg', GaussianProcessRegressor(kernel=kernel_gp, alpha=1e-2, n_restarts_optimizer=2, random_state=42))
])
```


ahora, continuamos con el entrenamiento para cada caso


```python
gp_x.fit(X_train_gp, y_train_gp.iloc[:, 0])
```

y para y : 

```python


gp_y.fit(X_train_gp, y_train_gp.iloc[:, 1])
```


realizamos una predicción en lotes (para controlar memoria)


```python


def predict_in_batches(model, X, batch_size=500):
    preds = []
    for i in range(0, len(X), batch_size):
        batch = X.iloc[i:i+batch_size]
        preds.append(model.predict(batch))
    return np.concatenate(preds)
```

de esta forma, obtenemos predicciones en validación y prueba:
