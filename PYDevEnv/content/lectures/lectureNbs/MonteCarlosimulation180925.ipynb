# MONTE CARLO SIMULATION 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python

#define n samples of monte carlo carlo simulation
M =  1000

N = 100 
A = 5 
noise = np.random.randn(N) # normal standar : mean = 0 , std = 1
x = A + noise 
 
```

```python
plt.plot(x.T, color = 'blue', alpha = 0.1) # nothing change to T... 
plt.plot(np.mean(x), color = 'red', linewidth = 2)
plt.show()
```

```python
plt.plot(x,label='x')
plt.plot(noise,label='noise')
plt.legend()
plt.show()
```

```python
# see mean and variance of x w
mx = np.mean(x)
mnoise = np.mean(noise) # cut ais = 1 
vw = np.var(x)
print(mx,mnoise,vw)
```

```python
plt.hist(noise,bins = 20) # check density = True funciton 
```

```python
# NOw i can implement the MONTE CARLO iteration
# # THIS IS A STOCHASTIC Process  
A1 = np.zeros((M,1))
A2= np.zeros((M,1))
for m in range(M): 
    print(f'Simulation {m+1}/{M} de monte Carlo')
    noise = np.random.randn(N) # normal standar : mean = 0 , std = 1
    x = A + noise
    A1[m] = np.mean(x) # this is my first estimator verosimilitud por Gaussiana
    A2[m]= x[0] # Promedio 2= muestra 1 # this is my second estimator 
    #print(A1)

plt.plot(noise,label='noise')
plt.plot(x,label='x')
plt.legend()
plt.show()  
```

```python
plt.hist(A1,bins = 20, alpha = 0.5,label='A1')
plt.hist(A2,bins = 20, alpha = 0.5,label='A2') # el menor es A1, el mayor es el mismo PRIMER ELEMENTO # this represent a lot of the variance
plt.legend()
plt.show() # legend and show to avoid text 
```
